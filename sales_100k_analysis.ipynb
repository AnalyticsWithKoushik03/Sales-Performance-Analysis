```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sales_100k.csv")
```

```python
df.head()
```

```python
category_wise_total_sales = df.groupby("Category")["Sales_Amount"].sum().astype(int).sort_values(ascending=False)
print(category_wise_total_sales)
```

```python
category_wise_total_profit = df.groupby("Category")["Profit"].sum().astype(int).sort_values(ascending=False)
print(category_wise_total_profit)
```

```python
total_quantity_of_category = df.groupby("Category")["Quantity"].sum().sort_values(ascending=False)
print(total_quantity_of_category)
```

```python
avg_sales_by_category = df.groupby("Category")["Sales_Amount"].mean().astype(int).sort_values(ascending=False)
print(avg_sales_by_category)
```

```python
avg_profit_of_category = df.groupby("Category")["Profit"].mean().astype(int).sort_values(ascending=False)
print(avg_profit_of_category)
```

```python
product_wise_total_sales = df.groupby("Product")["Sales_Amount"].sum().sort_values(ascending=False).head(10).astype(int)
print(product_wise_total_sales)
```

```python
total_profit_of_product = df.groupby("Product")["Profit"].sum().sort_values(ascending=False).head(10).astype(int)
print(total_profit_of_product)
```

```python
total_quantity_of_product = df.groupby("Product")["Quantity"].sum().sort_values(ascending=False).head(10)
print(total_quantity_of_product)
```

```python
avg_sales_of_product = df.groupby("Product")["Sales_Amount"].mean().sort_values(ascending=False).head(10).astype(int)
print(avg_sales_of_product)
```

```python
avg_profit_of_product = df.groupby("Product")["Profit"].mean().sort_values(ascending=False).head(10).astype(int)
print(avg_profit_of_product)
```

```python
Region_wise_total_sales = df.groupby("Region")["Sales_Amount"].sum().sort_values(ascending=False).astype(int)
print(Region_wise_total_sales)
```

```python
Region_wise_total_profit = df.groupby("Region")["Profit"].sum().sort_values(ascending=False).astype(int)
print(Region_wise_total_profit)
```

```python
Region_wise_total_quantity = df.groupby("Region")["Quantity"].sum().sort_values(ascending=False)
print(Region_wise_total_quantity)
```

```python
avg_sales_of_Region = df.groupby("Region")["Sales_Amount"].mean().sort_values(ascending=False).astype(int)
print(avg_sales_of_Region)
```

```python
avg_profit_of_Region = df.groupby("Region")["Profit"].mean().sort_values(ascending=False).astype(int)
print(avg_profit_of_Region)
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])
```

```python
monthly_sales = df.groupby(df["Order_Date"].dt.to_period("M"))["Sales_Amount"].sum().sort_values(ascending=False).head(10).astype(int)
print(monthly_sales)
```

```python
product_wise_total_sales.plot(kind="barh")
plt.title(" Product Wise Total Sales")
plt.xlabel("Sales")
plt.ylabel("Product")
```

```python
total_profit_of_product.plot(kind="barh")
plt.title("Total Profit Of Product")
plt.xlabel("Profit")
plt.ylabel("Product")
```

```python
avg_sales_of_product.plot(kind="barh")
plt.title("Average Sales of Product")
plt.xlabel("Sales")
plt.ylabel("Product")
```

```python
avg_profit_of_product.plot(kind="barh")
plt.title("Average Profit Of Product")
plt.xlabel("Profit")
plt.ylabel("Product")
```

```python
category_wise_total_sales.plot(kind="bar")
plt.title("Category Wise Total Sales")
plt.xlabel("Category")
plt.ylabel("Sales")
```

```python
category_wise_total_profit.plot(kind="bar")
plt.title(" Category Wise Total Profit")
plt.xlabel("Category")
plt.ylabel("Profit")
```

```python
avg_sales_by_category.plot(kind="bar")
plt.title("Average Sales By Category")
plt.xlabel("Category")
plt.ylabel("Sales")
```

```python
avg_profit_of_category.plot(kind="bar")
plt.title("Average Profit Of Category")
plt.xlabel("Category")
plt.ylabel("Profit")
```

```python
Region_wise_total_sales.plot(kind="pie",autopct="%1.1f%%")
plt.title("Region Wise Total Sales")
```

```python
Region_wise_total_profit.plot(kind="pie",autopct="%1.1f%%")
plt.title("Region Wise Total profit")
```

```python
avg_sales_of_Region.plot(kind="pie",autopct="%1.1f%%")
plt.title("Average Sales Of Region")
```

```python
avg_profit_of_Region.plot(kind="pie",autopct="%1.1f%%")
plt.title("Average Profit Of Region")
```

```python
monthly_sales.index = monthly_sales.index.astype(str)
monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
```
